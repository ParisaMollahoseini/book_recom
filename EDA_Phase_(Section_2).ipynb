```python
book_pca = pd.DataFrame(pca.transform(book[cols]), columns=[x for x in range(1,7)])
book_pca.head()
```

### âœ… Make rate distribution balance

```python
def Resampling(rate, proportion):
    global result
    n = ( proportion * (len(result)) - len(result[result['rating'] == rate]) ) / (1 - proportion)
    samp = result[result['rating'] == rate].sample(n=round(n), replace=True)
    balanced_result = pd.concat([result, samp], axis=0)
    result = balanced_result.copy()
```

```python
Resampling(1, 0.12)
Resampling(2, 0.12)
```

```python
print(result['rating'].value_counts() / result['rating'].value_counts().sum() * 100)
```

```python
result.to_csv(path + 'balanced_classifier_input.csv', index=False)
```
