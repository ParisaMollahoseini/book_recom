```python
book = pd.read_csv(path + 'final_book.csv').drop(['Unnamed: 0'],axis=1)
book.head()
```

```python
print(list(book.columns))
```

<div style="direction:rtl", align="justify">
<h3>
<span style='font-family:"B Nazanin"'>
پیش از انجام pca ستون های عددی را نرمالایز می کنیم. 

```python
from sklearn.preprocessing import normalize
cols = ['original_publication_year','average_rating','ratings_count','work_ratings_count',
        'work_text_reviews_count','ratings_1','ratings_2','ratings_3','ratings_4','ratings_5']

features = book[cols]
book[cols] = pd.DataFrame(normalize(book[cols]),columns = cols)
```

<div style="direction:rtl", align="justify">
<h3>
<span style='font-family:"B Nazanin"'>
در این قسمت میزان پوشش کامپوننت ها ضمن pca با کامپوننت های 1 الی 9 را بررسی می کنیم. اگر به pca با 6 کامپوننت توجه کنیم 99.6 درصد از تغییرات خروجی توسط 6 کامپوننت پوشش داده شده است که بسیار عدد مناسبی می باشد و با انتخاب آن  تعداد ستون های عددی از 10 ستون به 6 ستون کاهش پیدا می کند و هم پوشش خوبی خواهیم داشت.

```python
import numpy as np
from sklearn.decomposition import PCA

for i in range(1, 10):
    pca = PCA(n_components=i)
    pca.fit_transform(book[cols])
    print(i, np.cumsum(pca.explained_variance_ratio_))
    print('____________________________________________________________________')
```

```python
pca = PCA(n_components=6)
pca.fit_transform(book[cols])
print(np.cumsum(pca.explained_variance_ratio_))
```

```python
book_pca = pd.DataFrame(pca.transform(book[cols]), columns=[x for x in range(1,7)])
book_pca.head()
```

<div style="direction:rtl", align="justify">
<h3>
<span style='font-family:"B Nazanin"'>
ستون های flag مانند را با 6 ستون حاصل از pca جوین می کنیم.

```python
book_final = pd.concat([book_pca,book.iloc[:,11:]], axis=1)
book_final['book_id'] = book.book_id
book_final.head()
```

### ✅ Merge book and rating dataset

```python
ratings = pd.read_csv(path + 'ratings.csv')
```

```python
result = pd.merge(ratings, book_final, how='inner', on='book_id')
```

```python
result.head()
```

```python
result.shape
```

```python
result.to_csv(path + 'classifier_input.csv', index=False)
```

## 📑 Resampling

### ✅ Check distribution of rating variable

```python
result = pd.read_csv(path + 'classifier_input.csv')
result.head()
```

<div style="direction:rtl", align="justify">
<h3>
<span style='font-family:"B Nazanin"'>
در این قسمت توزیع متغیر rating درون دیتاست بررسی شده است همانطور که مشخص است بیشترین درصد یعنی 35 درصد از امتیازات امتیاز 4 بوده است و دو امتیاز 1 و 2 درصد بسیار پایینی دارند لذا برای اینکه مدل در یادگیری دچار خطا نشود و پیش بینی امتیازات پایین را نیز به خوبی یاد بگیرد با resample کردن درصد امتیازات 1 و 2 را افزایش می دهیم.

```python
result['rating'].value_counts()
```

```python
print(result['rating'].value_counts() / result['rating'].value_counts().sum() * 100)
```

### ✅ Make rate distribution balance

```python
def Resampling(rate, proportion):
    global result
    n = ( proportion * (len(result)) - len(result[result['rating'] == rate]) ) / (1 - proportion)
    samp = result[result['rating'] == rate].sample(n=round(n), replace=True)
    balanced_result = pd.concat([result, samp], axis=0)
    result = balanced_result.copy()
```

```python
Resampling(1, 0.12)
Resampling(2, 0.12)
```

```python
print(result['rating'].value_counts() / result['rating'].value_counts().sum() * 100)
```

```python
result.to_csv(path + 'balanced_classifier_input.csv', index=False)
```
