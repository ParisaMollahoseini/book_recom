⚫ Columns Description
<hr style="border:1px solid gray">
 00 🔷book_id : Book identifier in books dataset <br><br>
 01 🔷goodreads_book_id : Book identifier in original goodreads dataset <br><br>
 02 🔷best_book_id : Identifier which generally points to the most popular editions of a given book <br><br>
 03 🔷work_id : Refers to the book in the abstract sense <br><br>
 04 🔷books_count : The number of editions for a given work <br><br>
 05 🔷isbn : International Standard Book Number (ISBNs were 10 digits in length up to the end of December 2006) <br><br>
 06 🔷isbn13 : International Standard Book Number (since 1 January 2007 ISBNs now always consist of 13 digits) <br><br>
 07 🔷authors : Author or authors of the book <br><br>
 08 🔷original_publication_year : The year which book has been published <br><br>
 09 🔷original_title : The name under which the book was published <br>  <br>       
 10 🔷title : Original_title in short format<br>  <br>                   
 11 🔷language_code : The language of book <br><br>
 12 🔷average_rating : The average rating of the book received in total <br><br>
 13 🔷ratings_count : Total number of ratings the book received <br> <br>             
 14 🔷work_ratings_count : Total number of ratings the book received regarding all kinds of format or edition of each book <br>  <br>     
 15 🔷work_text_reviews_count : Total number of written text reviews the book received regarding all kinds of format or edition of each book <br><br> 
 16 🔷ratings_1 : Total number of rate (1) the book received <br>  <br>                
 17 🔷ratings_2 : Total number of rate (2) the book received <br> <br>             
 18 🔷ratings_3 : Total number of rate (3) the book received <br> <br>                
 19 🔷ratings_4 : Total number of rate (4) the book received <br> <br>                
 20 🔷ratings_5 : Total number of rate (5) the book received <br> <br>               
 21 🔷image_url : Url of the original image of book <br>   <br>
 22 🔷small_image_url : Url of the short image of book <br>
